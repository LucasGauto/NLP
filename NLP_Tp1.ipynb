## Ejercicio 2
El Ministerio de Turismo y Deportes de la Nación permite explorar tableros de información en línea tableros.yvera.tur.ar. Explore la página y utilizando una librería de web scraping extraiga los valores del tablero de indicadores de Objetivos de Desarrollo Sostenible en una tabla y el texto limpio de la metodología de los mismos.


```python
import requests
from bs4 import BeautifulSoup
```

```python
enlace = 'https://tableros.yvera.tur.ar/tablero_ODS/'
respuesta = requests.get(enlace, verify=False)
```

```python
soup = BeautifulSoup(respuesta.text, 'html.parser')
```

Tomo el html completo y me quedo solo con los divs que tienen la clase inner (son los que tienen los valores). Posteriormente se hace otra sopa con ese mismo texto y se busca cada elemento.

```python
resultado = [p for p in soup.find_all('div', class_='inner')]

type(resultado[0])
```

Asi se extraen los valores!!

```python
valores = []
for i in resultado:
    sopa = BeautifulSoup(str(i),'html.parser')
    valores.append(sopa.p.text)
print(valores)
```

Ahora faltan los textos...

```python
#Se usaran expresiones regulares para obtener aquellos textos que
#tengan el 2022 o la palabra Indice
import re
```

```python
todosLosParrafos = []
for i in resultado:
    sopa = BeautifulSoup(str(i),'html.parser')
    todosLosParrafos.append(p.text for p in sopa.find_all('p'))
```

```python
textos = []
for i in todosLosParrafos:
    for j in i:
        matches = re.search(r'.*\(2022\)', str(j))
        textos.append(matches)
```

```python
textos_final = []
for texto in textos:
    if texto != None:
        textos_final.append(texto.group())

textos_final = list(set(textos_final))
```

```python
for i in textos_final:
    print(i)
```

Ahora queda la creacion del dataframe y la carga de los datos

```python
import pandas as pd
```

```python
df = pd.DataFrame()
```

```python
df['Texto'] = [textos_final[5],
               textos_final[0],
               textos_final[2],
               textos_final[6],
               textos_final[3],
               textos_final[4],
               textos_final[1]]
df['Valores'] = valores
```

#### Convertir PDF en Imagenes

Para realizar esto se hará uso de la librería pdf2image.

```python
!pip install pdf2image
!apt-get install -y poppler-utils
```

```python
from pdf2image import convert_from_path
```

```python
#archivo = 'Metodos Estadisticos.pdf'

#Convertir en lista de imagenes
imagenes = convert_from_path('(INTRODUCTION TO THE THEORY OF STATISTICS) Alexander, Franklin Graybill and Duane Boes Mood - Introduccion a la teoria de la Estadistica-Aguilar (1969).pdf')

#Guardar imagenes
for i, imagen in enumerate(imagenes):
    imagen.save('pagina{}.png'.format(i), 'PNG')
```

Luego extraemos el texto con pytesseract

```python
!pip install pytesseract
!apt install tesseract-ocr
```

```python
from PIL import Image
import pytesseract

textoFinal = ''

for i in range(5):
  imagen = Image.open(f'pagina{i}.png')
  texto = pytesseract.image_to_string(imagen, lang="eng+spa")
  textoFinal+=texto
```
